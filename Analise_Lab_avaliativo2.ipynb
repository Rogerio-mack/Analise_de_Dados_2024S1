# **LAB A2 - Controle de Qualidade de Peças de uma Indústria**

**Pieces** é um data set para o controle de qualidade de peças produzidas por uma indústria. São empregadas quatro medidas (A, B, C e D) para o controle da qualidade das peças. As peças são então *Accept, Refurbish* ou *Reject* segundo o controle de qualidade (atributo `Quality`). A indústria ainda conta com 3 unidades para a produção dessas peças (atributo `Unit`).

Seu trabalho será de criar gráficos sobre esses dados.




# Aquisição dos Dados

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

df = pd.read_csv('http://meusite.mackenzie.br/rogerio/TIC/pieces3.csv')
```

```python
df.head()
```

# Exercício 1.

Crie uma figura figsize=(6,5) e com o modo artist layer crie 2x2 gráficos de densidade de probabilidade dos valores A, B, C e D, com a função `sns.kdeplot()`. Para o único valor que tem uma **distribuição mais próxima da normal, trace a linha do gráfico em vermelho**. Não deixe de nomear cada gráfico e os eixos.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = df.head()

fig, axes = plt.subplots(2, 2, figsize=(6, 5))

sns.kdeplot(data=df['A'], ax=axes[0, 0], color='green')
axes[0, 0].set_title('Distribuição de A')
axes[0, 0].set_xlabel('Valores')
axes[0, 0].set_ylabel('Densidade')

sns.kdeplot(data=df['B'], ax=axes[0, 1], color='darkblue')
axes[0, 1].set_title('Distribuição de B')
axes[0, 1].set_xlabel('Valores')
axes[0, 1].set_ylabel('Densidade')

sns.kdeplot(data=df['C'], ax=axes[1, 0], color='purple')
axes[1, 0].set_title('Distribuição de C')
axes[1, 0].set_xlabel('Valores')
axes[1, 0].set_ylabel('Densidade')

sns.kdeplot(data=df['D'], ax=axes[1, 1], color='pink')
axes[1, 1].set_title('Distribuição de D')
axes[1, 1].set_xlabel('Valores')
axes[1, 1].set_ylabel('Densidade')

sns.kdeplot(data=df['A'], ax=axes[0, 0], color='red', linestyle='--')

plt.tight_layout()

plt.show()
```

# Exercício 2.

Crie uma figura figsize=(6,3) e com o modo artist layer crie 1x2 gráficos de dispersão para os valores A $\times$ B e C $\times$ D. Para o gráfico A $\times$ B, inclua a dimensão C dos dados como o tamanho das 'bolhas' do gráfico. Para o gráfico C $\times$ D, inclua a dimensão `Unit` dos dados (você pode usar, por exemplo, cores diferentes). Posicione a legenda na parte inferior direita do gráfico. Não deixe de nomear cada gráfico e os eixos.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = df.head()

fig, axes = plt.subplots(1, 2, figsize=(9, 3))

sns.scatterplot(data=df, x='A', y='B', size='C', ax=axes[0], sizes=(20, 200))
axes[0].set_title('Gráfico A x B')
axes[0].set_xlabel('A')
axes[0].set_ylabel('B')

sns.scatterplot(data=df, x='C', y='D', hue='Unit', ax=axes[1], palette='viridis')
axes[1].set_title('Gráfico C x D')
axes[1].set_xlabel('C')
axes[1].set_ylabel('D')

axes[1].legend(loc='lower right')

plt.tight_layout()

plt.show()
```

# Exercício 3.

Crie uma figura figsize=(9,3) e com o modo artist layer crie 1x3 gráficos de barras com o seaborn para os valores Unit $\times$ B e Quality $\times$ B e Quality/Unit $\times$ B. No último gráfico elimine o intervalo de confiança dos valores. Posicione a legenda na parte inferior direita do gráfico. Não deixe de nomear cada gráfico e os eixos.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = df.head()

fig, axes = plt.subplots(1, 3, figsize=(9, 3))
fig.suptitle("Gráficos de Barras")

sns.barplot(data=data, x='Unit', y='B', ax=axes[0], ci=None)
axes[0].set_title("Unit x B")
axes[0].set_xlabel("Unit")
axes[0].set_ylabel("B")

sns.barplot(data=data, x='Quality', y='B', ax=axes[1], ci=None)
axes[1].set_title("Quality x B")
axes[1].set_xlabel("Quality")
axes[1].set_ylabel("B")

sns.barplot(data=data, x='Quality', y='B', hue='Unit', ax=axes[2], ci=None)
axes[2].set_title("Quality/Unit x B (sem IC)")
axes[2].set_xlabel("Quality/Unit")
axes[2].set_ylabel("B")
axes[2].legend(title='Unit', loc='lower right')

plt.tight_layout(rect=[0, 0, 1, 0.9])

plt.show()
```
